```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from keras.layers import Dense, Dropout
from keras.models import Sequential
from keras.optimizers import SGD,Adam
from keras.utils import to_categorical
```

```python
train = pd.read_csv('/content/fashion-mnist_train.csv')
test = pd.read_csv('/content/fashion-mnist_test.csv')
```

```python
print(train.shape)
print(test.shape)
```

```python
x_train = train.drop('label',axis=1)
y_train = train['label']

x_test = test.drop('label',axis=1)
y_test = test['label']
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
print(x_train)
```

```python
print(x_test)
```

Reshaping 2d into 3d images for better understanding

```python
x_train = x_train.values.reshape(len(x_train),28,28,1)
x_test = x_test.values.reshape(len(x_test),28,28,1)
```

```python
print(x_train)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

Data Augumentation - it actually reduces overfit of the model it will apply specified transformation with each image in batch

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
Gen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1
)
Gen.fit(x_train)
```

Building the model

```python
from tensorflow.keras.layers import Reshape, Flatten

model = Sequential()
model.add(Dense(128,activation='relu',input_shape=(28,28,1)))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
hist = model.fit(x_train,y_train,batch_size=128,epochs=30,verbose=1,validation_data=(x_test,y_test))
```

```python
test_acc = test_loss = model.evaluate(x_test,y_test,verbose=2)
print('Test Accuracy:',test_acc)
print('Test Loss:',test_loss)
```

```python
predictions = model.predict(x_test)
predictions_labels = np.argmax(predictions,axis=1)
```

```python
print(predictions_labels[:10])
print(np.argmax(y_test,axis=1)[:10])
```

```python
hist.history.keys()
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```
